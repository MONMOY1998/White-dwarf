```python
import numpy as np
import matplotlib.pyplot as plt
dxi=0.00001
N=1000000
m=np.arange(0,5.5,0.5)
for n in m:
    xi=0.01
    theta=1.0
    f1=0.0
    theta_sol=[]
    xi_sol=[]
    f1_sol=[]
    for i in range(N):
        f1+=-xi**2*theta**(n)*dxi
        theta+=f1/xi**2*dxi
        xi+=dxi
        theta_sol.append(theta)
        xi_sol.append(xi)
        f1_sol.append(f1)
        if(theta_sol[i]<0 and theta_sol[i-1]>=0):
             print (n,xi,f1)
        if(theta_sol[i]>0 and theta_sol[i-1]<=0):
            print(n,xi,f1)

    plt.plot(xi_sol,theta_sol)
plt.xlim(0,10)
plt.ylim(-0.4,1)
plt.axhline(y=0.0,color="black")
plt.xticks([0,2,4,6,8,10],size=14)
plt.yticks(size=14)
plt.tick_params(axis="x",length=6,width=1.5)
plt.tick_params(axis="y",length=6,width=1.5)
plt.minorticks_on()
plt.title("Solutions () to the Lane-Emden equation for various values of n",size=13,style="italic",weight="black",pad=15)
plt.xlabel("",size=15,weight="black")
plt.ylabel("()",size=15,weight="black")
plt.legend(["n=0","n=1","n=1.5","n=2","n=3","n=4","n=5","n=6"],markerfirst=False,title="Polytrope index",fontsize=8)
plt.tight_layout()
plt.show()
```
